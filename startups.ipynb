```python
#importar las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# leer el archivo .csv
df=pd.read_csv("50_Startups.csv")
```

```python
# 
X=df.iloc[ : ,:-1]
y=df.iloc[ : ,-1].values
y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# Transforma los datos de ciudad en datos numericos
ct=ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X=np.array(ct.fit_transform(X))
print(X)
```

```python
#importamos la libreria para seleccionar conjunto de pruebas
from sklearn.model_selection import train_test_split
#creamos las varian+bles x e y de pruba seleccionando que el 20% de los datos es el conjunto de prueba. Se escogen de manera aleatoria
print(X)
print(y)
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2)

```

```python
from sklearn.linear_model import LinearRegression
reggresor=LinearRegression()
reggresor.fit(X_train, y_train)
#Entrena el modelo, pra esto hay que asegurarse que este correctamente seleccionado los ejes X e y, ya que para este
#ejemplo solo se comparan numeros con numeros. No aparecen nombre de los estados.
```

```python
#Predecimos que nuestromodelo haya quedado correctamnete entrenado, haciendo predicciones sobre los
#valores x de prueba y nos deberia dar los mismos datos en el eje y que en la tabla y_test
y_pred=reggresor.predict(X_test)
dftest=pd.DataFrame({'Real values':y_test, 'predicted values': y_pred})
dftest
```

```python
# error cuadratico medio
from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

El anterior codigo nos devuelve el margen de error en comparando los datos de prueba con los datos de la prediccion
